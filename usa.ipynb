```python
import numpy as np
import pandas as pd
```

```python
usa = pd.read_html("https://developers.google.com/public-data/docs/canonical/states_csv")
usa_cord = pd.DataFrame(usa[0])
```

```python
usa_cord.head()
```

```python
del usa_cord['state']
```

```python
usa_cord.columns = ['latitude', 'longitude', 'state']
usa_cord.head()
```

```python
coronastats = pd.read_html('https://en.wikipedia.org/wiki/COVID-19_pandemic_in_the_United_States',match='U.S. state or territory')
```

```python
covid19 = pd.DataFrame(coronastats[0])

covid19.head()
```

```python
covid19 = covid19.iloc[:, :-4]
```

```python
covid19.head()
```

```python
covid19.columns = ['nana', 'state', 'total cases', 'deaths', 'recovered', 'hospitalized']
```

```python
del covid19['nana']
```

```python
del covid19['hospitalized']
```

```python
covid19.head()
```

```python
ds = pd.merge(covid19, usa_cord, how ='inner', on ='state')
ds.head()
```

```python
ds.info()
```

```python
import folium
```

```python
america = folium.Map(location = [37.778261,-100.417932], zoom_start = 4)
```

```python
for state,lat,long,total_cases,Death,Recov in zip(list(ds['state']),list(ds['latitude']),list(ds['longitude']),list(ds['total cases']),list(ds['deaths']),list(ds['recovered'])):
    
    folium.CircleMarker(location = [lat,long],
                       radius = 5,
                       color='red',
                       fill = True,
                       fill_color="red").add_to(america)
    
    folium.Marker(location = [lat,long],
                  # adding information that need to be displayed on popup
                  popup=folium.Popup(('<strong><b>State  : '+state+'</strong> <br>' +
                    '<strong><b>Total Cases : '+total_cases+'</striong><br>' +
                    '<strong><font color= red>Deaths : </font>'+Death+'</striong><br>' +
                    '<strong><font color=green>Recovered : </font>'+Recov+'</striong><br>'),max_width=200)).add_to(america)
    
```
